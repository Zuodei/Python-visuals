```python
import matplotlib.pyplot as plt

# Data
Yr = [2002, 2006, 2010, 2014, 2018, 2022]  # X-axis positions (Years)
GC = [6, 7, 5, 3, 4, 7]  # Goals Conceded
GS = [8, 5, 4, 7, 6, 11]  # Goals Scored

# Plot bars
plt.bar([y + 0.5 for y in Yr], GC, label="Goals Conceded", color="Red", width=0.6)
plt.bar([y - 0.5 for y in Yr], GS, label="Goals Scored", color="Blue", width=0.6)

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Number of Goals")
plt.legend()
plt.title("Argentina Goals Scored vs Goals Conceded in FIFA World Cups (2002-2022)")

# Automatically set xticks at the years
plt.xticks(Yr)  # Align ticks with the years
plt.yticks([]) #Remove y axis values

# Add text annotations for Goals Scored (GS)
for i, value in enumerate(GS):
    plt.text(Yr[i] - 0.5, value, str(value), ha='center', va='bottom', color='black')

# Add text annotations for Goals Conceded (GC)
for i, value in enumerate(GC):
    plt.text(Yr[i] + 0.5, value, str(value), ha='center', va='bottom', color='black')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the timeline
years = [2002, 2006, 2010, 2014, 2018, 2022]
top_scorers = ['Gabriel Batistuta & Hernan Crespo', 'Hernán Crespo & Maxi Rodriguez', 'Gonzalo Higuaín', 'Lionel Messi', 'Sergio Aquero', 'Lionel Messi']
goals = [1, 3, 4, 4, 2, 7]
top_assisters = ['Ariel Ortega', 'Juan Roman Riquelme', 'Lionel Messi', 'Ángel Di María', 'Lionel Messi', 'Lionel Messi']
assists = [1, 2, 3, 2, 2, 3]

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Create the timeline line
ax.hlines(y=1, xmin=min(years), xmax=max(years), color='gray', alpha=0.7, linewidth=2)

# Plotting the top scorers
ax.scatter(years, goals, color='blue', label='Top Scorers', zorder=5)
for i, txt in enumerate(top_scorers):
    ax.annotate(f'{txt} ({goals[i]} goals)', (years[i], goals[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9)

# Plotting the top assisters
ax.scatter(years, assists, color='red', label='Top Assisters', zorder=5)
for i, txt in enumerate(top_assisters):
    ax.annotate(f'{txt} ({assists[i]} assists)', (years[i], assists[i]), textcoords="offset points", xytext=(0, -10), ha='center', fontsize=9)

# Adding labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Number of Goals/Assists')
ax.set_title('Argentina National Team Top Scorers and Assisters in FIFA World Cups (2002-2022)')

# Adjust x-ticks to show all years and make labels readable
ax.set_xticks(years)
ax.set_xticklabels(years, rotation=45)

# Adding grid and legend
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend()

# Remove y-ticks as they are not needed for a linear timeline
ax.set_yticks([])

# Display the plot
plt.tight_layout()
plt.show()
```

```python
clsh = [1, 3, 2, 4, 0, 3] #List of clean sheets
glcn = [2, 2, 6, 4, 7, 8] #List of goals conceded
save = [7, 15, 20, 22, 12, 24] #List of saves
```

```python
plt.barh([y + 1 for y in Yr], clsh, color = "red", label = "Clean sheets")
plt.barh([y for y in Yr], glcn, color = "blue", label = "Goals conceded")
plt.barh([y - 1 for y in Yr], save, color = "green", label = "Saves")
plt.title("Argentina goalkeeper performance in FIFA world cups (2002-2022)")
plt.legend()
plt.yticks(Yr)

# Add text annotations for Clean Sheets (clsh)
for i, value in enumerate(clsh):
    plt.text(value + 0.1, Yr[i] + 1, str(value), va='center', ha='left', color='black')

# Add text annotations for Goals Conceded (glcn)
for i, value in enumerate(glcn):
    plt.text(value + 0.1, Yr[i], str(value), va='center', ha='left', color='black')

# Add text annotations for Saves (save)
for i, value in enumerate(save):
    plt.text(value + 0.1, Yr[i] - 1, str(value), va='center', ha='left', color='black')

plt.xticks([])
plt.show()
```

```python
Tack = [35, 55, 60, 75, 50, 80] #Number of tackles
Incp = [15, 25, 30, 40, 20, 45] #Number of interceptions
clrs = [20, 30, 35, 50, 30, 55] #Number of blocks and clearances
```

```python
plt.bar([y - 1 for y in Yr], Tack, label = "Tackles", color = "Red", width = 0.6)
plt.bar([y for y in Yr], Incp, label = "Interceptions", color = "Blue", width = 0.6)
plt.bar([y + 1 for y in Yr], clrs, label = "Blocks and Clearances", color = "Green", width = 0.6)
plt.legend()
plt.title("Argentina defensive performance in FIFA world cups (2002-2022)")
plt.xticks(Yr)

for i, value in enumerate(Tack):
    plt.text(Yr[i] - 1.4, value + 0.4, str(value))

for i, value in enumerate(Incp):
    plt.text(Yr[i] - 0.4, value + 0.4, str(value))

for i, value in enumerate(clrs):
    plt.text(Yr[i] + 0.6, value + 0.4, str(value))

plt.yticks([])
plt.show()
```
